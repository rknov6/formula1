```python
from pyspark.sql.functions import col
```

```python
dim_constructors_df = (
                       spark.table(f'{catalog}.{silver_schema}.constructors').alias("c")
                      .join(
                          spark.table(f'{catalog}.{gold_schema}.ref_nationality_region').alias("r")
                          ,"nationality","left")
                      .select(col("c.constructor_id"),col("c.constructor_name"),col("c.nationality"),col("r.region").alias("nationality_region"))
)
```

```python
dim_constructors_df.write.mode("overwrite").saveAsTable(f'{catalog}.{gold_schema}.dim_constructors')
```

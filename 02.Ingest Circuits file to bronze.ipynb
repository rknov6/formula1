```python
from pyspark.sql.functions import current_timestamp,col
```

```python
from pyspark.sql.types import StructField,StructType,StringType,DoubleType
```

```python
circuitschema = StructType([
                             StructField("circuitid",StringType()),
                             StructField("url",StringType()),
                             StructField("circuitName",StringType()),
                             StructField("lat",DoubleType()),
                             StructField("long",DoubleType()),
                             StructField("locality",StringType()),
                             StructField("country",StringType())
])
```

```python
df = spark.read.option("header","True").schema(circuitschema).csv("dbfs:/Volumes/formula1/landing/files/circuits.csv")
```

```python
 df = df.withColumn("filename", col("_metadata.file_name")) \
       .withColumn("ingestion_time", current_timestamp())
   
```

```python
df.write.mode("overwrite").saveAsTable("formula1.bronze.circuits")
```

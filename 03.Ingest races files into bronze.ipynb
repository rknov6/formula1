```python
from pyspark.sql.types import StringType,StructField,StructType,IntegerType,DoubleType,DateType
```

```python
from pyspark.sql.functions import col,current_timestamp
```

```python
race_schema = StructType([
                          StructField("season",IntegerType()),
                          StructField("round",IntegerType()),
                          StructField("url",StringType()),
                          StructField("raceName",StringType()),
                          StructField("date",DateType()),
                          StructField("circuitid",StringType())
                         ])
```

```python

(
    spark.read.option("header","true").option("mode","FAILFAST")
                .schema(race_schema)
                .csv("/Volumes/formula1/landing/files/races.csv")
                .withColumn("ingestion_time",current_timestamp())
                .withColumn("file_path",col("_metadata.file_path"))
                .write.mode("overwrite").format("delta").saveAsTable("formula1.bronze.races")
)
```

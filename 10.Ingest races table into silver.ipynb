```python
from pyspark.sql.functions import col,initcap
```

```python
%run ../formula1/00.Configurations
```

```python
renamed_columns = {"raceName": "race_name", "circuitid": "circuit_id", "date": "race_date"}
```

```python
races_df = (
             spark.table(f'{catalog}.{bronze_schema}.races')
                  .drop(col("url"))
                  .withColumnsRenamed(renamed_columns)
                  .dropDuplicates()
                  .withColumn("race_name",initcap((col("race_name"))))
)
```

```python
races_df.writeTo(f'{catalog}.{silver_schema}.races').createOrReplace()
```

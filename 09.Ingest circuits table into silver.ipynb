```python
from pyspark.sql.functions import col,initcap
```

```python
circuits_df = (
                 spark.table("formula1.bronze.circuits")
                      .drop(col("url"))
                      .withColumnsRenamed(
                           {"circuitid":"circuit_id",
                            "circuitName":"circuit_name",
                            "lat":"latitude",
                            "long":"longitude"}
                      )
                      .dropDuplicates(["circuit_id"])
                      .filter("circuit_id is not null")
                      .withColumn("circuit_name",initcap(col("circuit_name")))
                      .withColumn("locality",initcap(col("locality")))
)
circuits_df.display()
```

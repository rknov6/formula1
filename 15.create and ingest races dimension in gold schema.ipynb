```python
from pyspark.sql.functions import col
```

```python
circuits_df = spark.table(f'{catalog}.{silver_schema}.circuits')
```

```python
races_df = spark.table(f'{catalog}.{silver_schema}.races')
```

```python
final_df =(
           races_df.alias("r").join(circuits_df.alias("c"),"circuit_id","left")
                   .select(col("r.season"),col("r.round"),col("r.race_date"),col("r.race_name"),col("c.circuit_name"),col("c.locality"),col("c.country"))
          )
```

```python
final_df.write.mode("overwrite").saveAsTable(f'{catalog}.{gold_schema}.dim_races')
```

```python
from pyspark.sql.functions import col
```

```python
dim_drivers_df = (
                  spark.table(f'{catalog}.{silver_schema}.drivers').alias("d")
                  .join(spark.table(f'{catalog}.{gold_schema}.ref_nationality_region').alias("r"),"nationality","left")
                  .select(col("d.driver_id"),col("d.driver_name"),col("d.date_of_birth"),col("d.nationality"),col("r.region").alias("nationality_region"))
)
```

```python
dim_drivers_df.write.mode("overwrite").saveAsTable(f'{catalog}.{gold_schema}.dim_drivers')
```
